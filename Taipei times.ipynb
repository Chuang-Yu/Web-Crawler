```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
links = []
months1 = ['January','February','March','April','May','June','July','August','Septemper','October','November','December']
months2 = ['February','March','April','May','June','July','August','Septemper','October','November','December','January']
for j in range(18,20):
    for i in range(12):
        link = 'https://www.taipeitimes.com/News/list?section=all&reportrange='+months1[i]+'%201,%2020'+str(j)+'%20-%20'+months2[i]+'%201,%2020'+str(j)+'&keywords=trade%20war'
        if months2[i]=='January':
            link = 'https://www.taipeitimes.com/News/list?section=all&reportrange='+months1[i]+'%201,%2020'+str(j)+'%20-%20'+months2[i]+'%201,%2020'+str(int(j)+1)+'&keywords=trade%20war'
        print(j,months1[i],months2[i])
        print(link)
        links.append(link)
```

```python
news=[]
for link in links:
    res = requests.get(link)
    soup = BeautifulSoup(res.text,'html5lib')
    rows=soup.find_all('div', 'plistbox boxTitle boxText')
    for row in rows:
        r=row.find_all('a')
        for i in r:
            if 'href' in i.attrs:
                print(i['href'])
                news.append(i['href'])

```

```python
nn=[]
for i in news:
    new = {}
    res = requests.get(i)
    soup = BeautifulSoup(res.text, 'html5lib')
    rows = soup.find_all('div', 'page spage')
    new['發稿日']=rows[0].find('h6').text.strip()
    new['新聞標題']=rows[0].find('h1').text.strip()
    new['作者']=rows[0].find('div','name').text.strip()
    new['新聞內文']=rows[0].find_all('p')
    nn.append(new)
mo=pd.DataFrame.from_records(nn)
mo.to_excel('Taipei times.xlsx',index=False)
print(mo)
```
