```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
links = []
for i in range(1,110):
        link = 'https://nypost.com/search/trade+war/page/'+str(i)+'/'
        links.append(link)
        print(link)
```

```python
news=[]
for link in links:
    res = requests.get(link)
    soup = BeautifulSoup(res.text,'html5lib')
    rows=soup.find_all('ul','article-loop')[0].find_all('li')
    for row in rows:
        r=row.find_all('h3','entry-heading')[0].a['href']
        print(r)
        news.append(r)
     
```

```python
nn=[]
for i in news:
    new = {}
    res = requests.get(i)
    soup = BeautifulSoup(res.text, 'html5lib')
    rows = soup.find_all('div', 'content-wrapper ui-panel-animate ui-panel-content-wrap-closed ui-panel-content-wrap ui-panel-content-wrap-closed')
    new['發稿日']=rows[0].find('p','byline-date').text.strip().split('|')[0]
    try :
        article_title=rows[0].find('h1',).text.strip()
    except:
        article_title='None'
    new['新聞標題']=article_title
    try :
        article_author=(soup.find('div',id='author-byline').find_all('p')[0].text.strip()).replace('By','').split('and')
    except:
        article_author='None'
    new['作者']= article_author
    try :
        content=(rows[0].find_all('div','entry-content entry-content-read-more')[0]).find_all('p')
        new['新聞內文']=[(i.text.strip())for i in content] 
    except:
        new['新聞內文']='None'      
    new['網址']=i
    #print(new)
    nn.append(new)
mo=pd.DataFrame.from_records(nn)
mo.to_excel('New York Post.xlsx',index=False)
print(mo)
```
